**Objective:-**

The client a superstore chain wants to predict the likelihood of the customer giving a positive response and wants to identify the different factors which affect the customer’s response. 
We need to analyze the data provided to identify these factors and then build a prediction mode to predict the probability of a customer willing to give a positive response to the marketing campaign planned by the superstore chain for its upcoming offer.
